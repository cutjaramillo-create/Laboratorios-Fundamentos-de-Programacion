# Palabras reservadas

Las palabras reservadas (KeyWords) corresponden a los nombres de las declaraciones que el interprete de python incluye por defecto. No se deben utilizar estas palabras para asignar nombres a otros objetos.

El listado de palabras reservadas puede ser consultado ingresando el comando help("keyword") desde el interfaz.

**Ejemplo:**

```python
help("keywords")
```

# El espacio de nombres(namespaces ).
El espacio de nombres(namespaces ) contiene una relacion de los objetos existentes en la memoria del sistema y los nombres a los que estan ligados.


*   Python es un lenguaje de alto nivel en el que todos sus elementos son objetos invluyerndo los tipos de datos basicos de tal forma que los terminos "valor" y "objeto" son sinomimos.
*   Si un objeto no esta lugado al menos a un nombre, dicho objeto es desechado por el interprete.
*   Un objeto puede tener mas de un nombre.

# El operador de asignacion =

Para relacionar a un nombre con un objeto, se utiliza el operador = con la siguiente sintaxis:

<nombre> = <objeto>


**Ejemplos:**










```python
saludo = "Hola"
print(saludo)
num1 = 5
print(num1)
num2 = "6"
print(num2)
num3= num1 + 6
print(num3)
num4 = 54.23
print(num4)
```
